```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cdf = pd.read_csv("..\Clean Data\clean_data.csv")
cdf
```

```python
# Get Summary Statistics
cdf.describe()
```

```python
# Plot histogram for each pollutant
for col in cdf.columns[1:]:  # Skipping the date column
    plt.figure(figsize=(8, 3))
    sns.histplot(cdf[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Correlation Matrix
corr_mat = cdf.drop('Date' , axis = 1).corr()

# Correlation Analysis Using Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_mat, annot = True, cmap = 'coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Pollutants')
plt.show()
```

```python
# Creating Box plot for each columne to identify the presence of outliers and distribution of values
for col in cdf.columns[1:]:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=cdf[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# First let's aggregate data by months over Dates by calculating means for similar dates to make calculation easy
cdf['Date'] = pd.to_datetime(cdf['Date'])

# Aggregate by month
cdf_monthly = cdf.resample('M', on='Date').mean()

cdf_monthly
```

```python
# Plotting Line graph for each pollutant to analyze its trend over time
for col in cdf_monthly.columns[1:]: # Skipping date column
    plt.figure(figsize = (6,3))
    sns.lineplot(x = 'Date' , y = col , data = cdf_monthly)
    plt.title(f"{col} over time")
    plt.xticks(rotation = 45)
    plt.show()
```
